```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\LENOVO\IPL_Cricket_Analysis\IPL_Data.csv")
```

```python
# View first 5 rows
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
print("Duplicate rows removed successfully!")
```

```python
df.isnull().sum()
```

```python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

print("Missing values filled successfully!")
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
print("Remaining rows:", len(df))
```

```python
df.isnull().sum()
```

```python
df.to_csv("Cleaned_IPL_Data.csv", index=False)
print("Cleaned dataset saved successfully!")
```

```python
df.describe()
```

```python
summary = df.describe()
summary.to_csv("Summary_Report.csv")

print("Summary report saved successfully!")
```

```python
with pd.ExcelWriter("IPL_Report.xlsx") as writer:
    df.to_excel(writer, sheet_name="Cleaned_Data", index=False)
    summary.to_excel(writer, sheet_name="Summary")

print("Excel report created successfully!")
```
